```python
import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine

# Database credentials
db_user = 'postgres'
db_password = 'password'
db_host = '4.tcp.ngrok.io'
db_port = '12897'
db_name = 'dbdm'
table_name = 'nyc_req_full'

# Create database connection
DATABASE_URL = f"postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}"
engine = create_engine(DATABASE_URL)

# Query the table into DataFrame
query = f"SELECT * FROM {table_name};"
df = pd.read_sql(query, engine)

# Convert datetime columns
df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], errors='coerce')

# Calculate response time in hours
df['Response Time (hrs)'] = (df['Closed Date'] - df['Created Date']).dt.total_seconds() / 3600
df = df[(df['Response Time (hrs)'] > 0) & df['Response Time (hrs)'].notna()]

# Filter top 10 complaint types
top_complaints = df['Complaint Type'].value_counts().nlargest(10).index
df_filtered = df[df['Complaint Type'].isin(top_complaints)]

# Boxplot
fig_box = px.box(
    df_filtered,
    x="Complaint Type",
    y="Response Time (hrs)",
    title="Boxplot of Service Response Time by Complaint Type",
    labels={"Response Time (hrs)": "Response Time (Hours)"},
    points="outliers"
)
fig_box.update_layout(xaxis_tickangle=-45)
fig_box.show()

# Violin plot
fig_violin = px.violin(
    df_filtered,
    x="Complaint Type",
    y="Response Time (hrs)",
    title="Violin Plot of Service Response Time by Complaint Type",
    box=True,
    points="all",
    color="Complaint Type"
)
fig_violin.update_layout(xaxis_tickangle=-45, showlegend=False)
fig_violin.show()
```
